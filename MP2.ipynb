The dataset I am going to be focusing on is the NBA Player Stats for the middle of the 2022-2023 year.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
playersDF = pd.read_csv("nba.csv")

playersDF.head()
```

```python
cleanedPlayerDF = playersDF.dropna()

cleanedPlayerDF.head()
```

```python
birthGroup = playersDF.groupby("Age").size()

birthGroup
```

```python
fig, axs = plt.subplots(1, figsize=(5, 5))

positionGroup = cleanedPlayerDF.groupby('Pos').size()
plt.rcParams["font.size"] = 18

axs.pie(x=positionGroup, 
           labels=positionGroup.index, 
           autopct = '%1.0f%%')
plt.title("Percentage of Player's Position")


plt.show()
```

VISUAL 1: PERCENTAGE OF PLAYERS POSITIONS

```python
usDF = cleanedPlayerDF.loc[cleanedPlayerDF["Birth_Place"] == "us"]
notusDF = cleanedPlayerDF.loc[cleanedPlayerDF["Birth_Place"] != "us"]

len(usDF)

```

```python
len(notusDF)
```

```python
### Sourced some help from W3Schools!
y = np.array([len(usDF), len(notusDF)])
mylabels = ["US", "Other"]
myColors = ['purple', 'green']
plt.pie(y, labels = mylabels, autopct = '%1.0f%%',
           colors = myColors)
plt.title("Percentage of NBA Players from/not from the US")

plt.show() 
```

VISUAL 4: FIELD GOAL PERCENTAGES BY EACH POSITION

```python
pgDF = cleanedPlayerDF.loc[cleanedPlayerDF["Pos"] == "PG"]
sgDF = cleanedPlayerDF.loc[cleanedPlayerDF["Pos"] == "SG"]
sfDF = cleanedPlayerDF.loc[cleanedPlayerDF["Pos"] == "SF"]
pfDF = cleanedPlayerDF.loc[cleanedPlayerDF["Pos"] == "PF"]
cDF = cleanedPlayerDF.loc[cleanedPlayerDF["Pos"] == "C"]



```

```python


plt.hist(pgDF['FG%'], 
         alpha=0.5,
         label = 'Point Guard',
         color= '#d496a7')

plt.hist(sgDF['FG%'], 
         alpha=0.5,
         label = 'Shooting Guard',
         color= '#9d695a')

plt.hist(sfDF['FG%'], 
         alpha=0.5,
         label = 'Small Forward',
         color= '#78e0dc')

plt.hist(pfDF['FG%'], 
         alpha=0.5,
         label = 'Power Forward',
         color= '#a1cdf1')

plt.hist(cDF['FG%'], 
         alpha=0.5,
         label = 'Center',
         color= '#190933')

plt.legend(loc = 'upper left', fontsize = 10)
plt.xlabel('Field Goal %')
plt.ylabel('# of people')



plt.show()
```
